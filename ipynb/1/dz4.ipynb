```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./kc_house_data.csv')
df
```

### 1.2
Изучите стоимости недвижимости

```python
plt.figure(figsize=(8, 6))
plt.hist(df['price'], bins=80)
plt.title('Стоимости недвижимости')
plt.xlabel('Стоимость')
plt.ylabel('Количество');
```

### 1.3
Изучите распределение квадратуры жилой

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['sqft_living'], bins=70)
plt.title('Распределение квадратуры жилой')
plt.xlabel('Кв.м')
plt.ylabel('Количество');
```

### 1.4
Изучите распределение года постройки

```python
plt.figure(figsize=(8, 6))
plt.hist(df['yr_built'], bins=50)
plt.title('Распределение года постройки')
plt.xlabel('Год постройки')
plt.ylabel('Количество');
```

## Task 2

### 2.1
Изучите распределение домов от наличия вида на набережную
Постройте график
Сделайте выводы

```python
water = df['waterfront'].value_counts()
```

```python
plt.figure(figsize=(5, 5))
plt.pie(water, autopct='%1.1f%%')
plt.title('Распределение домов от наличия вида на набережную')
plt.legend(['no', 'yes']);
```

### 2.2
Изучите распределение этажей домов

```python
fl = df['floors'].value_counts()
```

```python
plt.figure(figsize=(5, 5))
plt.pie(fl, autopct='%1.1f%%', labels=fl.index)
plt.title('Количество этажей');
```

```python
plt.figure(figsize=(5, 5))
sns.barplot(x=fl.index, y=fl)
plt.title('Количество этажей')
plt.xlabel('Этажность')
plt.ylabel('Количество');
```

### 2.3
Изучите распределение состояния домов

```python
cond = df['condition'].value_counts()
```

```python
plt.figure(figsize=(5, 5))
plt.pie(cond, autopct='%1.1f%%', labels=cond.index)
plt.title('Распределение состояния домов');
```

```python
plt.figure(figsize=(5, 5))
sns.barplot(x=cond.index, y=cond)
plt.title('Распределение состояния домов')
plt.xlabel('Оценка состояния')
plt.ylabel('Количество');
```

```python
corr_matrix = df.corr(numeric_only=True)
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
```
